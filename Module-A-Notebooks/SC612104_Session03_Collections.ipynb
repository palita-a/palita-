## 1. ทำไมต้องมี Collection?

จนถึงตอนนี้ ตัวแปรที่เราใช้เก็บได้แค่ **1 ค่าต่อตัวแปร** เช่น `name = "มานี"` เก็บชื่อได้คนเดียว

ลองคิดปัญหานี้: ถ้าต้องเก็บชื่อนักศึกษาทั้งห้อง 40 คน จะสร้างตัวแปร 40 ตัวไหม?

```python
# วิธีแย่มาก - ไม่ทำตามนี้!

student3 = "วิชัย"
```

**Collection** คือชนิดข้อมูลที่เก็บ "หลายค่า" ไว้ในตัวแปรเดียว Python มี 4 collection หลักที่ต้องรู้จัก:

| Collection | เก็บข้อมูลแบบ | เปลี่ยนแปลงได้ไหม | มีลำดับไหม | ซ้ำกันได้ไหม |
|---|---|---|---|---|
| `list` | ลำดับของค่า | ✅ แก้ไขได้ (mutable) | ✅ มีลำดับ | ✅ ซ้ำได้ |
| `tuple` | ลำดับของค่า | ❌ แก้ไขไม่ได้ (immutable) | ✅ มีลำดับ | ✅ ซ้ำได้ |
| `set` | กลุ่มของค่าที่ไม่ซ้ำ | ✅ แก้ไขได้ | ❌ ไม่มีลำดับ | ❌ ไม่ซ้ำ |
| `dict` | คู่ key-value | ✅ แก้ไขได้ | ✅ มีลำดับ (Python 3.7+) | key ไม่ซ้ำ, value ซ้ำได้ |

จะเห็นความแตกต่างชัดเจนขึ้นเรื่อย ๆ ในแต่ละหัวข้อด้านล่าง

### 2.2 การเข้าถึงสมาชิกใน list (Indexing)

ทุกสมาชิกใน list มี "ตำแหน่ง" หรือ index กำกับอยู่ — **Python เริ่มนับ index จาก 0 เสมอ** ไม่ใช่ 1!

```
index:    0       1       2       3
list:  ["แอปเปิ้ล", "กล้วย", "ส้ม", "มะม่วง"]
neg-idx: -4      -3      -2      -1
```

นอกจากนับจากซ้าย (0, 1, 2, ...) ยังนับจากขวาด้วย index ลบได้ (-1 คือตัวสุดท้าย)

````python
print(fruits[0])    # แอปเปิ้ล (ตัวแรก)
print(fruits[2])    # ส้ม
print(fruits[-1])   # มะม่วง (ตัวสุดท้าย)
print(fruits[-2])   # ส้ม (ตัวรองสุดท้าย)

# ถ้า index เกินขอบเขต จะเกิด error
# print(fruits[10])   # ลบ # แล้วรันดู จะเห็น IndexError
````

| `.append(x)` | เพิ่ม x ไปท้าย list | `fruits.append("ฝรั่ง")` |
| `.insert(i, x)` | แทรก x ที่ตำแหน่ง i | `fruits.insert(1, "ฝรั่ง")` |
| `.remove(x)` | ลบค่า x ตัวแรกที่เจอ | `fruits.remove("กล้วย")` |

| `x in list` | เช็คว่า x อยู่ใน list ไหม | `"กล้วย" in fruits` |


````python
fruits = ["แอปเปิ้ล", "กล้วย", "ส้ม"]
````

````python
fruits.append("มะม่วง")
print("หลัง append:", fruits)
````

## 3. tuple

**tuple** คือ collection ที่เก็บค่าหลายค่าเรียงเป็นลำดับ **เหมือน list ทุกอย่าง ยกเว้นแก้ไขไม่ได้** (immutable) ใช้เครื่องหมาย `( )` ในการสร้าง

### 3.1 การสร้าง tuple


````python
# สร้าง tuple
coordinates = (13.736717, 100.523186)   # ละติจูด, ลองจิจูด ของกรุงเทพฯ
rgb_color = (255, 0, 1)                  # สีแดงในรูปแบบ RGB

print(coordinates)
print(rgb_color)
print(type(coordinates))
````

````python
rgb_color[2]
````

````python
# tuple ที่มีสมาชิกตัวเดียว ต้องมี , ต่อท้าย ไม่งั้นจะไม่ใช่ tuple!
not_a_tuple = (5)        # นี่คือ int ธรรมดา ไม่ใช่ tuple
single_tuple = (5,)      # ต้องมี comma ต่อท้าย จึงเป็น tuple
print(type(not_a_tuple))   # <class 'int'>
print(type(single_tuple))  # <class 'tuple'>
````

### 3.2 การเข้าถึงสมาชิกใน tuple

ใช้ index และ slicing แบบเดียวกับ list ได้ทุกอย่าง — ต่างกันแค่**แก้ไขค่าไม่ได้**


````python
coordinates = (13.736717, 100.523186)

print(coordinates[0])    # 13.736717
print(coordinates[1])    # 100.523186

# ลองแก้ไขค่า -> จะเกิด error เพราะ tuple แก้ไขไม่ได้
# coordinates[0] = 14.0   # ลบ # แล้วรันดู จะเห็น TypeError
````

### 3.3 Tuple Unpacking — แตกค่าจาก tuple ใส่ตัวแปรหลายตัว

เป็นการใช้งานที่นิยมมากของ tuple — เอาแต่ละค่าของ tuple ไปใส่ตัวแปรในบรรทัดเดียว


````python
coordinates = (13.736717, 100.523186)

latitude, longitude = coordinates   # unpack ทั้ง 2 ค่าในบรรทัดเดียว
print(f"ละติจูด: {latitude}, ลองจิจูด: {longitude}")
````

````python
# ใช้บ่อยมากตอน return ค่าหลายค่าจากฟังก์ชัน (function จะเรียนละเอียดในคาบถัดไป)
def get_min_max(numbers):
    return min(numbers), max(numbers)   # จริงๆ คือ return tuple (min, max)

result = get_min_max([5, 2, 8, 1, 9])
print(result)          # (1, 9) -- ได้ tuple กลับมา
low, high = result      # unpack ได้เลย
print(f"ต่ำสุด: {low}, สูงสุด: {high}")
````

````python
minv, maxv = get_min_max([2,5,8,1])
print(minv, maxv)
````

### 3.4 เมื่อไหร่ใช้ tuple แทน list?

| ใช้ `list` เมื่อ... | ใช้ `tuple` เมื่อ... |
|---|---|
| ข้อมูลต้องเปลี่ยนแปลงได้ (เพิ่ม/ลบ/แก้) | ข้อมูลไม่ควรถูกแก้ไขเลย (เช่น พิกัด, ค่าคงที่) |
| ลำดับรายการที่ขยาย/หดได้ | กลุ่มค่าที่มีจำนวนคงที่และความหมายชัดเจนตามตำแหน่ง |
| เช่น รายชื่อนักศึกษาที่เพิ่ม/ลบได้ | เช่น (วัน, เดือน, ปี) หรือ (R, G, B) |

**ข้อดีเสริม:** tuple ทำงานเร็วกว่า list เล็กน้อย และใช้เป็น key ของ dict ได้ (list ใช้เป็น key ไม่ได้ เพราะแก้ไขได้)


````python
# ตัวอย่าง: ใช้ tuple เป็น key ของ dict (list ทำแบบนี้ไม่ได้!)
distances = {
    (13.7, 100.5): "กรุงเทพฯ",
    (18.8, 98.9): "เชียงใหม่",
}
print(distances[(13.7, 100.5)])
````

## 4. set

**set** คือ collection ที่เก็บค่าที่ **ไม่ซ้ำกัน** และ **ไม่มีลำดับ** ใช้เครื่องหมาย `{ }` ในการสร้าง (ต้องมีสมาชิกข้างใน ไม่งั้นจะกลายเป็น dict ว่าง)

### 4.1 การสร้าง set


````python
# สร้าง set
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม", "แอปเปิ้ล"}   # ใส่ "แอปเปิ้ล" ซ้ำ 2 ครั้ง
print(fruits_set)   # จะเหลือแค่ตัวเดียว เพราะ set ไม่เก็บค่าซ้ำ
print(type(fruits_set))
````

````python
# สร้าง set ว่าง - ต้องใช้ set() ไม่ใช่ {} (เพราะ {} คือ dict ว่าง)
empty_set = set()
print(type(empty_set))
````

````python
x = {}
type(x)
````

````python
# แปลง list เป็น set เพื่อลบค่าซ้ำ (ใช้บ่อยมาก!)
numbers_with_duplicates = [1, 2, 2, 3, 3, 3, 4]
unique_numbers = set(numbers_with_duplicates)
print(unique_numbers)   # {1, 2, 3, 4}
````

### 4.2 การเพิ่ม/ลบสมาชิกใน set

**ข้อสำคัญ:** set ไม่มีลำดับ จึงเข้าถึงสมาชิกผ่าน index แบบ `fruits_set[0]` ไม่ได้ (จะ error) ต้องใช้ loop หรือเช็คด้วย `in` แทน


````python
fruits_set = {"แอปเปิ้ล", "กล้วย", "ส้ม"}
````

````python
fruits_set.add("มะม่วง")          # เพิ่มสมาชิก
print("หลัง add:", fruits_set)
````

````python
fruits_set.remove("กล้วย")        # ลบสมาชิก (error ถ้าไม่มีอยู่จริง)
print("หลัง remove:", fruits_set)
````

````python
fruits_set.discard("ทุเรียน")     # ลบสมาชิก (ไม่ error แม้ไม่มีอยู่จริง - ปลอดภัยกว่า remove)
print("หลัง discard (ไม่มีทุเรียนอยู่แล้ว แต่ไม่ error):", fruits_set)
````

````python
print("มีส้มไหม:", "ส้ม" in fruits_set)
print("จำนวนสมาชิก:", len(fruits_set))
````

### 4.3 Set Operations — การคำนวณทางคณิตศาสตร์ของเซต

จุดเด่นที่สุดของ set คือการคำนวณแบบเซตในคณิตศาสตร์ (Union, Intersection, Difference) ซึ่งมีประโยชน์มากในการวิเคราะห์ข้อมูล เช่น หาความเหมือน/ต่างของ 2 กลุ่ม

| Operation | สัญลักษณ์ | ความหมาย | function |
|---|---|---|---|
| Union | \| | รวมสมาชิกทั้งสองเซต (ไม่ซ้ำ) | `.union()` |
| Intersection | & | สมาชิกที่มีอยู่ใน**ทั้งสอง**เซต | `.intersection()` |
| Difference | - | สมาชิกที่มีใน A แต่ไม่มีใน B | `.difference()` |
| Symmetric Difference | ^ | สมาชิกที่อยู่ในเซตใดเซตหนึ่งเท่านั้น (ไม่ใช่ทั้งคู่) | `.symmetric_difference()` |


````python
# ตัวอย่าง: นักศึกษาที่ลงทะเบียนวิชา A และวิชา B
students_course_a = {"มานี", "สมชาย", "วิชัย", "ปรีชา"}
students_course_b = {"สมชาย", "วิชัย", "สมหญิง", "อรุณ"}

# Union: นักศึกษาทั้งหมดที่ลงเรียนวิชาใดวิชาหนึ่ง (หรือทั้งสอง)
print("Union:", students_course_a | students_course_b)
students_course_a.union(students_course_b)
````

````python
# Intersection: นักศึกษาที่ลงทั้ง 2 วิชา
print("Intersection:", students_course_a & students_course_b)
````

````python
# Difference: นักศึกษาที่ลงวิชา A แต่ไม่ได้ลงวิชา B
print("Difference (A-B):", students_course_a - students_course_b)
````

````python
# Symmetric Difference: นักศึกษาที่ลงแค่วิชาเดียว (ไม่ใช่ทั้งคู่)
print("Symmetric Difference:", students_course_a ^ students_course_b)
students_course_a.symmetric_difference(students_course_b)
````

### 4.4 เมื่อไหร่ใช้ set?

- ต้องการ**ลบค่าซ้ำ**ออกจากข้อมูล (วิธีที่เร็วและง่ายที่สุด)

## 6. สรุปเปรียบเทียบ list vs tuple vs set vs dict

| | `list` | `tuple` | `set` | `dict` |
|---|---|---|---|---|
| **สัญลักษณ์สร้าง** | `[1, 2, 3]` | `(1, 2, 3)` | `{1, 2, 3}` | `{"a": 1, "b": 2}` |
| **แก้ไขได้ (mutable)** | ✅ | ❌ | ✅ | ✅ |
| **มีลำดับ (ordered)** | ✅ | ✅ | ❌ | ✅ (key ตามลำดับที่ใส่) |
| **มีค่าซ้ำได้** | ✅ | ✅ | ❌ | key ไม่ซ้ำ, value ซ้ำได้ |
| **เข้าถึงผ่าน** | index `[0]` | index `[0]` | ไม่มี index ตรง ใช้ `in`/loop | key `["ชื่อ"]` |
| **ใช้เมื่อ** | ข้อมูลที่ต้องแก้ไข/ขยาย | ข้อมูลคงที่ ไม่ควรแก้ | ต้องการค่าไม่ซ้ำ/set operations | ข้อมูลที่มีชื่อกำกับ (key-value) |

### ตัวอย่างข้อมูลเดียวกัน เก็บด้วย collection คนละแบบ


````python
# ข้อมูลเดียวกัน (คะแนนสอบของนักศึกษา) เก็บด้วย collection ต่างๆ กัน

# list: เก็บคะแนนตามลำดับการสอบ (ลำดับสำคัญ และอาจมีคะแนนซ้ำกัน)
scores_list = [85, 90, 85, 78, 92]

# tuple: เก็บค่าคงที่ เช่น ช่วงคะแนนของเกรด A (ไม่ควรเปลี่ยน)
grade_a_range = (80, 100)

# set: เก็บคะแนนที่ไม่ซ้ำ (เผื่ออยากรู้ว่ามีคะแนนกี่แบบที่แตกต่างกัน)
unique_scores = set(scores_list)

# dict: เก็บคะแนนพร้อมชื่อนักศึกษา (มีชื่อกำกับแต่ละค่า)
scores_dict = {"มานี": 85, "สมชาย": 90, "วิชัย": 78}

print("list:", scores_list)
print("tuple:", grade_a_range)
print("set (ค่าที่ไม่ซ้ำ):", unique_scores)
print("dict:", scores_dict)
````

### ข้อ: การจัดการข้อมูลสินค้า
คุณมีข้อมูลสินค้าดังนี้:
*   `all_products = ["เสื้อ", "กางเกง", "รองเท้า", "กระเป๋า", "เสื้อ", "หมวก"]` (รายการสินค้าทั้งหมด, มีสินค้าซ้ำ)
*   `in_stock = {"เสื้อ", "กางเกง", "รองเท้า"}` (สินค้าที่มีในสต็อก)

จงเขียนโค้ดเพื่อ:
1.  สร้าง `unique_products` ที่เป็น `set` จาก `all_products` เพื่อให้ได้รายการสินค้าที่ไม่ซ้ำกันทั้งหมด
2.  หาว่ามีสินค้าอะไรบ้างที่**หมดสต็อก** (อยู่ใน `unique_products` แต่ไม่อยู่ใน `in_stock`) แล้วเก็บในตัวแปร `out_of_stock`.
3.  สร้าง `product_prices` เป็น `dict` โดยมี `key` เป็นชื่อสินค้าและ `value` เป็นราคา กำหนดราคาเองสำหรับสินค้า 3 ชิ้นแรกใน `unique_products` (เช่น `{"เสื้อ": 350, "กางเกง": 500, "รองเท้า": 800}`)
4.  ดึงราคาของ `"เสื้อ"` ออกมาเก็บในตัวแปร `shirt_price` โดยใช้ `.get()`

````python
# Print ผลลัพธ์แต่ละข้อ

all_products = ["เสื้อ", "กางเกง", "รองเท้า", "กระเป๋า", "เสื้อ", "หมวก"]
in_stock = {"เสื้อ", "กางเกง", "รองเท้า"}

# 1. สร้าง unique_products
unique_products = set(all_products)
print(f"Unique products: {unique_products}")


# 2. หาสินค้าหมดสต็อก
out_of_stock = unique_products - in_stock
print(f"Out of stock: {out_of_stock}")



# 3. สร้าง product_prices
# หมายเหตุ: เนื่องจาก set ไม่มีลำดับตายตัว การเลือก '3 ชิ้นแรก' อาจได้ผลต่างกันในแต่ละรัน
# เพื่อความแน่นอน จะใช้ค่าจาก unique_products ที่ถูกแปลงเป็น list ก่อน หรือกำหนดเองเลย
# Corrected: product_prices should be a dictionary with product names as keys and prices as values.
product_prices = {"เสื้อ": 199, "กระเป๋า": 299, "กางเกง": 289}
print(f"Product prices: {product_prices}")



# 4. ดึงราคาเสื้อ
# Retrieve 'เสื้อ' price from the product_prices dictionary using .get()
shirt_price = product_prices.get("เสื้อ")
print(f"Shirt price: {shirt_price}")
````

### ข้อ: การจัดการข้อมูลนักศึกษาในชั้นเรียน

คุณมีข้อมูลนักศึกษาในชั้นเรียนดังนี้:
`class_records = { "S001": {"name": "มานี", "scores": {"Math": 85, "Science": 90}}, "S002": {"name": "สมชาย", "scores": {"Math": 70, "Science": 75}} }`

จงเขียนโค้ดเพื่อ:
1.  **เพิ่มนักศึกษาใหม่:** เพิ่มนักศึกษาชื่อ "วิชัย" รหัส "S003" ด้วยคะแนน `{"Math": 92, "Science": 88}` เข้าไปใน `class_records`
2.  **อัปเดตคะแนน:** อัปเดตคะแนนวิชา "Science" ของนักศึกษา "S001" เป็น 95
3.  **คำนวณคะแนนเฉลี่ย:** คำนวณคะแนนเฉลี่ยของนักศึกษา "S002" และเก็บในตัวแปร `avg_score_S002` (ผลลัพธ์ควรเป็น float)
4.  **หานักเรียนที่ได้คะแนนเฉลี่ยสูงสุด:** หานักเรียนที่ได้คะแนนเฉลี่ยรวมทุกวิชาสูงสุด และ print ชื่อของนักเรียนคนนั้นออกมา (อนุญาตให้ใช้ `max()` ร่วมกับ `key` หรือ `dict comprehension` แต่ไม่ใช้ `for` loop แบบปกติในการวนหาค่าสูงสุด)

````python
class_records = {
    "S001": {"name": "มานี", "scores": {"Math": 85, "Science": 90}},
    "S002": {"name": "สมชาย", "scores": {"Math": 70, "Science": 75}}
}

print("ข้อมูลเริ่มต้น:", class_records)

# 1. เพิ่มนักศึกษาใหม่
class_records["S003"]={"name":" จิตตรี","scores":{"Math":92},"Science":88}
print("เพิ่มนักศึกษาใหม่:",class_records["S003"])



# 2. อัปเดตคะแนน
class_records["S001"]["scores"]["Science"]=95
print("อัปเดตคะแนน:",class_records["S001"])



# 3. คำนวณคะแนนเฉลี่ยของ S002
# ใช้ sum() และ len() กับค่าใน dict.values()
avg_score_S002 = (class_records["S002"]["scores"]["Math"]+class_records["S002"]["scores"]["Science"])/2
print("คะแนนเฉลี่ย S002:",float(avg_score_S002))



# 4. หานักเรียนที่ได้คะแนนเฉลี่ยสูงสุด (ใช้ max ร่วมกับ key)
# สร้าง dictionary ของคะแนนเฉลี่ยสำหรับแต่ละคนก่อน (dict comprehension)
avg_scores = {student:sum(data["scores"].values())/len(data["scores"])for student,data in class_records.items()}
max_student = max(avg_scores,key=avg_scores.get)

print("นักเรียนที่ได้คะแนนเฉลี่ยสูงสุด:",class_records[max_student]["name"])
# ใช้ max() กับ student_averages.items() โดยกำหนด key ให้หาค่าสูงสุดจาก value
````
